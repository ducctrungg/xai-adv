## **XGBoost**

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
percentages = ["20%", "40%", "60%", "80%", "100%"]
lightgbm = [0.7565,0.98063,0.75675,0.7565, 0.75656]
random_forest = [0.2169,0.71714,0.30184,0.67214, 0.32644]
cnn_2 = [0.35409,0.35366,0.39851,0.47202, 0.49650]
cnn_4 = [0.35463,0.35491,0.35556,0.35561, 0.50644]

plt.figure(figsize=(10, 6))

# Plotting the data
plt.plot(percentages, lightgbm, 'o-', label='LightGBM', color='orange')
plt.plot(percentages, random_forest, 'v-', label='Random Forest', color='blue')
plt.plot(percentages, cnn_2, 's-', label='CNN 2 lớp', color='cyan')
plt.plot(percentages, cnn_4, 'p-', label='CNN 4 lớp', color='red')

# Adding title and labels
plt.title('Sử dụng mô hình thay thế XGBoost tấn công các mô hình IDS')
plt.xlabel('Tỉ lệ phần trăm số mẫu dữ liệu được sử dụng')
plt.ylabel('Độ chính xác của mô hình')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```
